# Clustering data points

In this assigmnent you'll show me which algorithms can be used to cluster some data points that have no known label.

We start, as usual, by importing some modules.

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_blobs
```

And we generate our dataset

```
# Generate some random cluster data
X1, y = make_blobs(random_state=170, n_samples=600, centers = 5)
rng = np.random.RandomState(74)

# Transform the data to be stretched
transformation = rng.normal(size=(2, 2))
X = np.dot(X1, transformation)

# Plot
# plt.scatter(X1[:, 0], X1[:, 1])
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

This is a fonction to show how your model performs.

```
def show_model_prediction(model, X):
  labels = model.fit_predict(X)
  plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=30, cmap='viridis');
```

### Clustering Algorithm #1

Please describe the algorithm you intended to use and tell me why you chose it.

**Answer: Agglomerative clustering also known as hierarchical clustering, I chose it because it's unsupervised clustering method that grows partitions by merging points together   **

```
from sklearn.cluster import AgglomerativeClustering
# Instantiate and fit your model here
model_1 =AgglomerativeClustering(n_clusters=5)
model_1.fit(X)
show_model_prediction(model_1, X)
```

### Clustering Algorithm #2

Please describe the algorithm you intended to use and tell me why you chose it.

**Answer: Kmeans is a unsupervised learning clustering algorithm, therefore suitable in this problem case **

```
from sklearn.cluster import KMeans
# Instantiate and fit your model here
model_2 = KMeans(n_clusters=5, random_state=0)
model_2.fit(X)
show_model_prediction(model_2, X)
```
